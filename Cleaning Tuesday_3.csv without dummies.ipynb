## Based on above, will delete Appellation_4 (since all are US) and rename Appellation 3 as STATE

```python
df = df.drop(columns=['Appellation4'])
```

```python
df.Appellation2.value_counts(normalize=False, sort=True, ascending=True, bins=None, dropna=False)
```

```python
df.Appellation2.fillna(df.Appellation1, inplace=True)
```

```python
df.Appellation3.fillna(df.Appellation2, inplace=True)
```

```python
df.Appellation3.value_counts(normalize=False, sort=True, ascending=True, bins=None, dropna=False)
```

```python
states = ["New York","Washington","Oregon","California"]
for s in states:
    df.loc[df['Appellation3'].str.contains(s), 'State'] = s
```

```python
df.State.value_counts(normalize=False, sort=True, ascending=True, bins=None, dropna=False)
```

```python
df = df.rename(columns={'Appellation2': 'Broad_Region'}) 
df = df.rename(columns={'Region': 'Broad_Region'}) 
```

```python
df = df.rename(columns={'Appellation1': 'Town'}) 
```

```python
df.describe(include= 'all')
```

```python
df.Region2.value_counts(normalize=False, sort=True, ascending=True, bins=None, dropna=False)
```

```python
df= df.drop(columns=['Region2','Broad_Region','Appellation3','Town'])
```

```python
df= df.drop(columns=['Variety'])
```

```python
df = df.rename(columns={'Var': 'Variety'}) 
```
