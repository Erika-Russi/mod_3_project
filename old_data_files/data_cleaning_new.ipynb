```python
import pandas as pd
```

```python
df = pd.read_csv('tuesday_1.csv')
df = df.drop(columns='Unnamed: 0')
```

```python
df.head()
```

```python
### KEEP APPELLATION AND DO NLP BEFORE SPLIT
```

```python
#split appellation
new1 = df['Appellation'].str.split(",", n = 4, expand = True)
df['Appellation1'] = new1[0]
df['Appellation2'] = new1[1] 
df['Appellation3'] = new1[2] 
df['Appellation4'] = new1[3] 
#df.drop(columns =['Appellation'], inplace = True)
df
```

```python
#drop '/A' from Price
df = df.drop(df[df.Price == '/A'].index)
```

```python
#remove $ and ", Buy Now" from price and change to a float
df.Price = df.Price.str.strip(to_strip = ', Buy Now')
df.Price = df.Price.str.strip(to_strip = '$')
df.Price = df.Price.astype(float)
```

```python
df.head()
```

```python
#remove percantage(%) from alcohol
df.Alcohol = df.Alcohol.str.strip(to_strip = '%')
df
```

```python
df.Wine
```

```python
#pulling year (must change the series name on left of =)
df.Bottlesize = df.Wine.str.extract(r'(?P<Year>\d{4})')
df.rename(columns={'Bottlesize': 'Year'})
```

```python
df.head()
```

```python
df.info()
```

```python
#dropping rows that do not have a review, winery, or alcohol content (very small amount of total set)
df = df.dropna(subset=['Review'])
df = df.dropna(subset=['Winery'])
df = df.dropna(subset=['Alcohol'])
```

```python
#changing alcohol values to numeric
df["Alcohol"] = pd.to_numeric(df["Alcohol"])
```

```python
#removing wines with alcohol content over 25% (three values above 40 - not wine)
df = df[df['Alcohol'] < 25]
```

```python
#pull varietal from 'Wine' and add to new column: wine_type
df['wine_type'] = df.Wine.str.extract(r'(?<=\d{4} )([A-Za-z ]+)(?= \(?)')
df.head()
```
